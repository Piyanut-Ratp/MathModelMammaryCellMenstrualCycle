```python
import numpy  as np
from matplotlib import pyplot as plt
```

P4 --> RANKL 

Rp4 = konrankl*[P4]/Kmrankl+[P4]

image.png

Wang, Jun et al. Science translational medicine(2013). https://doi.org/10.1126/scitranslmed.3006883


image.png

```python
k_onrankl= 0.0065
K_mrankl=163.09392
P4=50
#P4=20
rankl=0.031
P4t=[P4]
ranklt=[rankl]

t0=0
tend=24
h=1 #nonlinear less time is good. ## step side (less): number of population (more)
Time=[t0]

for i in np.arange(t0+h, tend+h, h):

  v1 = h*((k_onrankl*P4)/(K_mrankl+P4))

  #print(v1)

  P4=P4
  rankl=rankl+v1
  

  ranklt.append(rankl)
  P4t.append(P4)
  Time.append(i)

# plt.plot(cFost,E2ft)
plt.plot(Time, ranklt)
plt.xlabel("Time")
plt.ylabel("RANKL")
```
