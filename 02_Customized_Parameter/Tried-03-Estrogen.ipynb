```python
import numpy  as np
from matplotlib import pyplot as plt
```

# Estrogen to ERK

image.png

A. L. Scaling, E. R. Prossnitz, H. J. Hathaway. Hormones and Cancer(2014). http://doi.org/10.1007/s12672-014-0174-1

image.png

```python
k_onpERKE2=0.244
E2=10
pERK=0.13
#Et=[E2]
pERKt=[pERK]

t0=0
tend=0.25
h=0.01 #nonlinear less time is good. ## step side (less): number of population (more)
Time=[t0]

for i in np.arange(t0+h, tend+h, h):

  v1 = h*(k_onpERKE2*E2)

  #print(v1)

  pERK=pERK+v1
  

  pERKt.append(pERK)
  Time.append(i)

# plt.plot(cFost,E2ft)
plt.plot(Time, pERKt)
plt.xlabel("Time")
plt.ylabel("pERK")
plt.xlim([0,0.25])
```

# Estrogen to cytoC

image.png

Y. Deng, Y. Miki, A. Nakanishi. Scientific Reports(2020). http://doi.org/10.1038/s41598-020-57819-9

image.png

```python
k_onCytocE2=0.66
K_mCytocE2=4.36358
E2=32
Cytoc=0
#Et=[E2]
Cytoct=[Cytoc]

t0=0
tend=24
h=1 #nonlinear less time is good. ## step side (less): number of population (more)
Time=[t0]

for i in np.arange(t0+h, tend+h, h):

  v1 = h*((k_onCytocE2*E2)/(K_mCytocE2+E2))

  #print(v1)

  Cytoc=Cytoc+v1
  

  Cytoct.append(Cytoc)
  Time.append(i)

# plt.plot(cFost,E2ft)
plt.plot(Time, Cytoct)
plt.xlabel("Time")
plt.ylabel("CytoC")
#plt.xlim([0,0.25])
```
